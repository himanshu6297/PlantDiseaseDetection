```python
import os, shutil, random, re

# вњ… Your current dataset path (based on what I see in Jupyter)
SOURCE_TRAIN = r"Train"

# вњ… New output folder we will create
SEVERITY_BASE = r"PlantSeverity"
ALL_DIR = os.path.join(SEVERITY_BASE, "all")  # temporary bucket before splitting

# Clean old output if you rerun
if os.path.exists(SEVERITY_BASE):
    shutil.rmtree(SEVERITY_BASE)

os.makedirs(os.path.join(ALL_DIR, "Healthy"), exist_ok=True)
os.makedirs(os.path.join(ALL_DIR, "Mild"), exist_ok=True)
os.makedirs(os.path.join(ALL_DIR, "Severe"), exist_ok=True)

def severity_from_folder(folder_name: str) -> str:
    name = folder_name.lower()

    # Healthy
    if "healthy" in name:
        return "Healthy"

    # Severe (late, virus, greening, black rot, esca/black measles)
    severe_keywords = [
        "late", "virus", "yellow leaf curl", "mosaic",
        "huanglongbing", "citrus greening", "black rot",
        "esca", "black measles"
    ]
    if any(k in name for k in severe_keywords):
        return "Severe"

    # Everything else -> Mild
    return "Mild"

img_ext = (".jpg", ".jpeg", ".png", ".bmp", ".webp")
total = 0
counts = {"Healthy": 0, "Mild": 0, "Severe": 0}

for disease_folder in os.listdir(SOURCE_TRAIN):
    disease_path = os.path.join(SOURCE_TRAIN, disease_folder)
    if not os.path.isdir(disease_path):
        continue

    sev = severity_from_folder(disease_folder)

    for fn in os.listdir(disease_path):
        if not fn.lower().endswith(img_ext):
            continue

        # вњ… avoid overwriting (many folders can contain same filename)
        safe_folder = re.sub(r"[^A-Za-z0-9_ -]+", "_", disease_folder)
        new_name = f"{safe_folder}__{fn}"

        src = os.path.join(disease_path, fn)
        dst = os.path.join(ALL_DIR, sev, new_name)
        shutil.copy2(src, dst)

        counts[sev] += 1
        total += 1

print("вњ… Severity conversion done!")
print("Total images copied:", total)
print("Counts:", counts)
print("Output folder:", ALL_DIR)
```

```python
import os, shutil, random

ALL_DIR = r"PlantSeverity/all"
OUT_TRAIN = r"PlantSeverity/train"
OUT_VAL   = r"PlantSeverity/val"
OUT_TEST  = r"PlantSeverity/test"

# Clean old split folders if rerun
for p in [OUT_TRAIN, OUT_VAL, OUT_TEST]:
    if os.path.exists(p):
        shutil.rmtree(p)

classes = ["Healthy", "Mild", "Severe"]
random.seed(42)

for c in classes:
    os.makedirs(os.path.join(OUT_TRAIN, c), exist_ok=True)
    os.makedirs(os.path.join(OUT_VAL, c), exist_ok=True)
    os.makedirs(os.path.join(OUT_TEST, c), exist_ok=True)

def split_list(items, train_ratio=0.8, val_ratio=0.1):
    n = len(items)
    n_train = int(n * train_ratio)
    n_val = int(n * val_ratio)
    train_items = items[:n_train]
    val_items = items[n_train:n_train+n_val]
    test_items = items[n_train+n_val:]
    return train_items, val_items, test_items

for c in classes:
    files = [f for f in os.listdir(os.path.join(ALL_DIR, c)) if f.lower().endswith((".jpg",".jpeg",".png",".bmp",".webp"))]
    random.shuffle(files)

    train_files, val_files, test_files = split_list(files, 0.8, 0.1)

    for f in train_files:
        shutil.copy2(os.path.join(ALL_DIR, c, f), os.path.join(OUT_TRAIN, c, f))
    for f in val_files:
        shutil.copy2(os.path.join(ALL_DIR, c, f), os.path.join(OUT_VAL, c, f))
    for f in test_files:
        shutil.copy2(os.path.join(ALL_DIR, c, f), os.path.join(OUT_TEST, c, f))

    print(c, "-> train:", len(train_files), "val:", len(val_files), "test:", len(test_files))

print("\nвњ… Split complete!")
print("Train:", OUT_TRAIN)
print("Val  :", OUT_VAL)
print("Test :", OUT_TEST)
```

```python
import tensorflow as tf
from tensorflow.keras import layers

SEED = 42

train_ds = train_ds.cache().prefetch(AUTOTUNE)
val_ds = val_ds.cache().prefetch(AUTOTUNE)
test_ds = test_ds.cache().prefetch(AUTOTUNE)

TRAIN_DIR = r"PlantSeverity/train"
VAL_DIR   = r"PlantSeverity/val"
TEST_DIR  = r"PlantSeverity/test"

IMG_SIZE = (128, 128)   # smaller
BATCH = 4               # very small batch
EPOCHS = 5              # fewer epochs

# вњ… If you have GPU, this prevents TensorFlow from grabbing all GPU memory at once
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for g in gpus:
            tf.config.experimental.set_memory_growth(g, True)
        print("вњ… GPU memory growth enabled")
    except Exception as e:
        print("GPU memory growth error:", e)

train_ds = tf.keras.utils.image_dataset_from_directory(
    TRAIN_DIR, image_size=IMG_SIZE, batch_size=BATCH, seed=SEED, label_mode="int"
)
val_ds = tf.keras.utils.image_dataset_from_directory(
    VAL_DIR, image_size=IMG_SIZE, batch_size=BATCH, seed=SEED, label_mode="int"
)
test_ds = tf.keras.utils.image_dataset_from_directory(
    TEST_DIR, image_size=IMG_SIZE, batch_size=BATCH, shuffle=False, label_mode="int"
)

class_names = train_ds.class_names
print("Classes:", class_names)  # ['Healthy','Mild','Severe']

AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.shuffle(1000, seed=SEED)
val_ds = val_ds
test_ds = test_ds

# вњ… class weights (train counts you already confirmed)
train_counts = {"Healthy": 18833, "Mild": 19191, "Severe": 7801}
total = sum(train_counts.values())
class_weight = {
    0: total / (3 * train_counts["Healthy"]),
    1: total / (3 * train_counts["Mild"]),
    2: total / (3 * train_counts["Severe"]),
}
print("class_weight:", class_weight)

data_aug = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.08),
    layers.RandomZoom(0.1),
])

base = tf.keras.applications.MobileNetV2(
    include_top=False,
    input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3),
    weights="imagenet"
)
base.trainable = False

inputs = layers.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3))
x = data_aug(inputs)
x = tf.keras.applications.mobilenet_v2.preprocess_input(x)
x = base(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.2)(x)
outputs = layers.Dense(3, activation="softmax")(x)

model = tf.keras.Model(inputs, outputs)
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=["accuracy"]
)

callbacks = [
    tf.keras.callbacks.ModelCheckpoint("severity_best.keras", save_best_only=True, monitor="val_accuracy", mode="max"),
    tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True, monitor="val_accuracy"),
    tf.keras.callbacks.ReduceLROnPlateau(patience=2, factor=0.3, monitor="val_loss"),
]

print("\nрџљЂ Starting Training...\n")

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    callbacks=callbacks,
    class_weight=class_weight,
    verbose=1
)

print("\n Training Finished!\n")
```

```python
best = tf.keras.models.load_model("severity_best.keras")
test_loss, test_acc = best.evaluate(test_ds)
print("Final Test Accuracy:", test_acc)
```
